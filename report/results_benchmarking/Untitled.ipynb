```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
average_comm_avoid = []

average_our_result = []
file = pd.read_csv('result.txt', sep = "\t", header=0).values
for i in range(1,5): 
        results_comm_avoid = np.array(pd.read_csv("comm_avoiding_" + str(i) +".out").values[:,5], dtype=np.float32)
        average_comm_avoid += [np.average(results_comm_avoid)]
        average_our_result += [file[i][1]]

#average_our_result = file[:][1]) - For some reason, this does not work 
plt.ylabel("Time [s]")
plt.xlabel("Number of Threads")
plt.ylim(1,10)
plt.plot([1,2,3,4], average_comm_avoid)
plt.plot([1,2,3,4], average_our_result)
plt.legend(("Communication Avoiding", "Our parallel implementation"))
plt.savefig("benchmarking_0.001_USA_graph.png")
```
